```
from fastai import *
from fastai.vision import *
from fastai.metrics import error_rate
```

# Gather Data

```
from google.colab import files
def upload(path):
    uploaded = files.upload()
    with open(path,'wb') as fp:
        fp.write(uploaded[list(uploaded.keys())[0]])
```

```
upload("urls_tub.txt")
upload("urls_lawson.txt")
upload("urls_wingback.txt")
upload("urls_slipper.txt")
upload("urls_bergere.txt")
upload("urls_modern_wingback.txt")
upload("urls_club.txt")
upload("urls_occasional.txt")
```

```
classes = ["tub", "lawson", "wingback", "slipper", "bergere", "modern_wingback", "club", "occasional"]
```

```
path = Path('data/chairs')
dest = path/"tub"
dest.mkdir(parents=True, exist_ok=True)
```

```
download_images("urls_tub.txt", dest, max_pics=200)
```

```
for c in classes[1:]:
  dest = path/c
  dest.mkdir(parents=True, exist_ok=True)
  download_images("urls_" + c + ".txt", dest, max_pics=200)
```

```
for c in classes:
  print(c)
  verify_images(path/c, delete=True, max_workers=8)
```

```
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2, ds_tfms=get_transforms(), size=299, bs=32).normalize(imagenet_stats)
```

```
data.show_batch(rows=3, figsize=(7,8))
```

```
data.classes, data.c
```

# Training Model

```
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```
learn.fit_one_cycle(4)
```

**This is some terrible error rate we got here.**

```
learn.save("stage-1")
```

# Results

```
interp = ClassificationInterpretation.from_learner(learn)
losses, idxs = interp.top_losses()
len(data.valid_ds) == len(losses) == len(idxs)
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
interp.plot_confusion_matrix(figsize=(12, 12), dpi=60)
```

```
interp.most_confused(min_val=2)
```

# Retraining

```
learn.unfreeze()
```

```
learn.fit_one_cycle(1)
```

```
learn.load('stage-1')
```

```
learn.lr_find()
```

```
learn.recorder.plot()
```

```
learn.unfreeze()
learn.fit_one_cycle(2, max_lr=slice(1e-6, 1e-3))
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
interp.plot_confusion_matrix(figsize=(12, 12), dpi=60)
```

```
interp.most_confused(min_val=2)
```
